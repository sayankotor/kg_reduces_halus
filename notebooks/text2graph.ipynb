```python
import os, json
import pandas as pd
import glob

import os, json
import pandas as pd
import fnmatch

import re
files = glob.glob('/home/jovyan/shares/SR004.nfs2/chekalina/label_generation/*000.csv')
```

```python
from ast import literal_eval
from tqdm import tqdm
texts = []
entities = []
entity2emb = {}

char_window = 50
for file in tqdm(files):
    data = pd.read_csv(file)
    for i in tqdm(range(len(data))):
        item = data.iloc[i]
        entity_labels = literal_eval(item["entities"])
        entity_ids = literal_eval(item["entity_numbs"])
        embeddings = literal_eval(item["entity_embs"])
        text = item["text"]
        for entity, entity_id, embedding in zip(entity_labels, entity_ids, embeddings):
            entity2emb[entity_id] = embedding
            pos = [i for i in range(len(text)) if text.startswith(entity, i)]
            for p in pos:
                # print(p)
                texts.append(text[p - char_window: p + char_window])
                entities.append(entity_id)

                
```

# Training

```python
import torch
from transformers import AutoTokenizer, AutoModelForMaskedLM

DEVICE = torch.device("cuda:0")

name = "FacebookAI/roberta-large"
tokenizer = AutoTokenizer.from_pretrained(name)
model = AutoModelForMaskedLM.from_pretrained(name).to(DEVICE)
```

```python
import torch.nn as nn
TEXT_EMB_DIM = 1024
GRAPH_EMB_DIM = 200
projector = nn.Sequential(
    nn.Linear(TEXT_EMB_DIM, GRAPH_EMB_DIM),
    nn.LeakyReLU(),
    nn.Linear(GRAPH_EMB_DIM, GRAPH_EMB_DIM)
).to(DEVICE)
```

```python
optimizer = torch.optim.AdamW(list(model.parameters()) + list(projector.parameters()), lr=1e-4)
```

```python
from IPython.display import clear_output
import matplotlib.pyplot as plt
%matplotlib inline

grad_accum = 64
losses_train = []
cosines_train = []
# losses_test = []

N_ITERS = 10000
```

```python
import random
import numpy as np


losses_batch = []
for iters in range(N_ITERS):
    ind = random.randint(0,len(texts)-1)
    text = texts[ind]
    entity_id = entities[ind]
    embedding = torch.FloatTensor(entity2emb[entity_id]).to(DEVICE)
    input_ids = tokenizer.encode(text, return_tensors="pt").to(DEVICE)
    output = model(input_ids, output_hidden_states=True)
    pooled_emb = output["hidden_states"][-1].mean(1)
    projected_emb = projector(pooled_emb)
    
    loss = nn.MSELoss()(projected_emb, embedding)
    losses_batch.append(loss.item())
    loss.backward()
    
    if iters % grad_accum == 0:
        losses_train.append(np.mean(losses_batch))
        cosines_train.append(torch.cosine_similarity(embedding, projected_emb).item())
        losses_batch = []
        optimizer.step()
        optimizer.zero_grad()
        plt.title("losses_train")
        plt.semilogy(losses_train)
        plt.grid()
        plt.show()
        plt.title("cosines between predicted embedding and true embedding")
        plt.plot(cosines_train)
        plt.grid()
        plt.show()
        clear_output(True)
    
```

```python
def text2graph_emb(text):
    with torch.no_grad():
        input_ids = tokenizer.encode(text, return_tensors="pt").to(DEVICE)
        output = model(input_ids, output_hidden_states=True)
        pooled_emb = output["hidden_states"][-1].mean(1)
        projected_emb = projector(pooled_emb)
    
    return projected_emb
```
